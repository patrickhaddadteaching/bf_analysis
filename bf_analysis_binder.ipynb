```python
import numpy as np
from scipy.stats import binom
from ipywidgets import interact, interact_manual, widgets, Label
```

```python
nb_p=10000
v_p=0.5*np.arange(1,nb_p+1,dtype=np.float64)/(nb_p+1)
v_h=-1*(np.log2(v_p)*(v_p)+np.log2(1-v_p)*(1-v_p))
```

```python
#tests_results=widgets.HTML(value='',placeholder='',description='',disabled=True, continuous_update=True,layout=widgets.Layout(width='200px',height='300px'))
tests_results=widgets.HTML(value='',placeholder='',description='',disabled=True, continuous_update=True)
word_size=128

rv05=binom(word_size,0.5)
v_x=np.arange(word_size+1)
v_y05=rv05.pmf(v_x)
def smart_bf(entropy_in=1.0,sens_biais=[0,1],s_option=['unsorted','sorted','sorted and cumulated'],nb_col=1):
  
  if entropy_in==1:
    p_target=0.5
  elif entropy_in==0:
    p_target=int(sens_biais)
  else:
    rnd=sens_biais
    if rnd==0:
      p_target=v_p[np.where(v_h>entropy_in)[0][0]]
    else:
      p_target=1-v_p[np.where(v_h>entropy_in)[0][0]]
  rv=binom(word_size,p_target)
  
  v_y=rv.pmf(v_x)  
  if s_option=='unsorted':
    v_index_sorted0=v_x
    v_index_sorted1=v_x
    s_header='Chance to win'
    v_y_to_plot=v_y
    v_y05_to_plot=v_y05
  elif s_option=='sorted':
    v_index_sorted0=np.argsort(v_y)[::-1]
    v_index_sorted1=v_index_sorted0
    s_header='Chance to win'
    v_y_to_plot=v_y
    v_y05_to_plot=v_y05
  else:
    v_index_sorted0=np.argsort(v_y)[::-1]
    v_index_sorted1=v_x
    v_y_to_plot=(v_y[v_index_sorted0]).cumsum()
    v_y05_to_plot=(v_y05[v_index_sorted0]).cumsum()
    s_header='Cumulative Chance'
  """
  s_label_test=''
  for i in range(v_index_sorted0.shape[0]):
    s_label_test='%s <p>%d %f %f</p>'%(s_label_test,v_index_sorted0[i],v_y_to_plot[v_index_sorted1[i]],np.log2(v_y05_to_plot[v_index_sorted1[i]])+word_size)
  """
  
  s_label_test='<table>'
  nb_lines=int(word_size/nb_col)+1
  cnt_elem=0
  for i_col in range(nb_col):
    s_label_test='%s<td>HW</sup></td><td>&nbsp&nbsp</td><td>&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbspGroup Size</td><td>&nbsp&nbsp</td><td>&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp%s</td><td>&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp</td>'%(s_label_test,s_header)
  for i_lines in range(nb_lines):
    s_label_test='%s<tr>'%s_label_test
    for i_col in range(nb_col):
      if cnt_elem<(word_size+1):
          s_label_test='%s<td>%d</sup></td><td>&nbsp&nbsp</td><td>&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp2<sup>%f</sup>&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp</td><td>&nbsp&nbsp</td><td>%.10e</td><td>&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp</td>'%(s_label_test,v_index_sorted0[cnt_elem],np.log2(v_y05_to_plot[v_index_sorted1[cnt_elem]])+word_size,v_y_to_plot[v_index_sorted1[cnt_elem]]*100)
      cnt_elem+=1
    s_label_test='%s</tr>'%s_label_test
  tests_results.value=s_label_test
  #plt.plot(v_x,v_y)
  #plt.show()
im=interact(smart_bf)
im.widget.close()
im.widget.children[0].max=1.0
im.widget.children[0].min=0.0
im.widget.children[0].step=0.01
im.widget.children[0].description='Entropy'
im.widget.children[1].description='Bias bit value'
im.widget.children[2].description='Strategy'
im.widget.children[3].min=1
im.widget.children[3].max=3
im.widget.children[3].description='Display'
v_box_top=widgets.VBox([im.widget.children[0],im.widget.children[1],im.widget.children[2],im.widget.children[3],tests_results])
display(v_box_top)
```
